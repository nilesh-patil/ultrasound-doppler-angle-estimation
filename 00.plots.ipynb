```python
%pylab inline

import pandas as pd
import seaborn as sns
```

```python
results_30 = pd.read_csv('./03.results/combined_output_30.txt',sep='|')
results_60 = pd.read_csv('./03.results/combined_output_60.txt',sep='|')
```

```python
results_30['delta'] = results_30.y_actual-results_30.y_predicted
results_60['delta'] = results_60.y_actual-results_60.y_predicted
```

```python
results_30['condition'] = results_30.image_rotations.between(-31,31)
results_60['condition'] = results_30.image_rotations.between(-61,61)
```

```python
sns.set(style="whitegrid", palette="pastel", color_codes=True, rc={'figure.figsize':(16,6)})

sns.boxplot(x="model", y="delta", hue="condition", data=results_60, palette={False: "b", True: "y"} )
plt.xticks(rotation=60);
```

```python
results = pd.concat([results_30, results_60])

results = results.reset_index()

results.head(2)
```

```python
results['condition'] = (
                        results.image_rotations.between(-30,30)
                        )

results['ModelType'] = results.model.str[-2:]
```

```python
results.head(2)
```

```python
results['pointSize'] = (results.delta-results.delta.min())/(results.delta.max()-results.delta.min())
```

```python
import matplotlib.patches as mpatches
import matplotlib.pyplot as plt

red_patch = mpatches.Patch(color='red', label='Model - 30')
blue_patch = mpatches.Patch(color='blue', label='Model - 60')
```

```python
sns.set_style('white')

plt.figure(figsize=(10,10), dpi=300)

plt.scatter(x = results.image_rotations,
            y = results.delta,
            s = 10*results.pointSize,
            c=results.ModelType.apply(lambda row: 'blue' if row=='60' else 'red'),
            alpha=0.05
           )
plt.xlabel('Rots')
plt.ylabel('Delta')

plt.legend(handles=[red_patch, blue_patch])
plt.grid(True);
```
